# Leer archivo .parquet de errores

```python
# logicData_etl/utils/read_quarantine_errors_pyarrow.py
import argparse
import pandas as pd
```

```python
def main():
    parser = argparse.ArgumentParser()
    parser.add_argument('--path', required=False, help='s3://bucket/... or local file/folder')
    ns = parser.parse_args()
    path = ns.path

    if path.startswith("s3://"):
        # requiere s3fs instalado y credenciales AWS configuradas
        df = pd.read_parquet(path, engine="pyarrow", storage_options={})
    else:
        df = pd.read_parquet(path, engine="pyarrow")

    print("shape:", df.shape)
    print(df.head(50))

```

```python
if __name__ == "__main__":
    main()
```
